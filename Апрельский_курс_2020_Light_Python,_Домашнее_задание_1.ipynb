### Задача 5

```python
# Условия задачи были интерпретированы так, что сложить необходимо как начальное (предыдущее) число со следующим, так и сами числа в диапазоне между собой
n = 15 # создаем переменную
i = [n] # создаем список с единственным элементом для последующего добавления значений от 16 до 22 для их суммирования
while n < 22: # создаем цикл, который будет повторяется пока n не будет равна 22
    c = n # создаем переменную для запоминания числа 15 и его вывода в фразе "Сумма .. и .. равна ...". Если не сделать эту переменную, то значение будет n будет всегда +1 от начального из-за строки 18
    a = n + 1 # создаем переменную для хранения числа на 1 больше предыдущего
    b = a + n # создаем переменную для хранения суммы предыдущего и следующего числа
    n = n + 1 # увеличиваем переменную n на +1 для следующей итерации
    i.append(a) # добавляем в список значение n увеличенного на +1 для сохранения диапазона чисел от 15 до 22
    print('Сумма', c, 'и', a,'равна', b) # выводим сумму предыдущего числа и следующего числа
print('Сумма всех челых чисел от 15 до 22 включительно равна',sum(i)) # выводим сумму чисел 15, 16, 17, 18, 19, 20, 21, 22
```

Используя цикл (любой), найдите значение максимального элемента списка из предыдущего задания (без использования встроенных функций max и т.д.). Можно использовать встроенные методы списков (любые). 

P.S. Не стесняйтесь гуглить :)

P.S. 2 Гуглить нужно фразу "алгоритм поиска максимального элемента массива".

Пример: для списка [3, 4, 1, 7, 2] значением максимального элемента является число 7.

```python
'''
Решение №1. Задачу можно решить при помощи встроенного метода my_list.sort() и ссылки на последний элемент списка.
Однако в данном решении нет использования цикла и сравнения числа с другими,
чтобы сделать логический вывод о максимальности значения этого числа
'''
my_list = [56, 23, 67, 45, 67, 2, 47, 158, 31, 34, 78, 23, 78, 23, 89, 23, 36]
my_list.sort()
print('Решение №1. Максимальное значение элемента списка:', my_list[-1])
```

```python
'''
Решение №2. 
Алгоритм: поиск необходимого значания методом исключения. 
Сравниваем значения между собой, удаляем наименьшее, оставляя наибольшее.
Реализация: При помощи цикла while c использованием метода my_list.pop() и без использования метода my_list.sort()
'''
my_list = [56, 23, 67, 45, 67, 2, 47, 158, 31, 34, 78, 23, 78, 23, 89, 23, 36]
try:
    while True:
        while my_list[1] <= my_list[0]:
            my_list.pop(1)
        while my_list[1] >= my_list[0]:
            my_list.pop(0)
except:
    print('Решение №2. Максимальное значение элемента списка:', my_list[0])
```

```python
'''
Решение №3. 
Алгоритм: Находим нужное значение методом исключения,
сравнивая элементы листа друг с другом по-очереди.
При нахождении наименьшего удаляем его из списка.
Механика: При помощи цикла for, методов my_list.append(), my_list.index(), my_list.pop()
Примечание: добавление элемента i обусловлено тем, что если его не добавить переменная element_index может получить значение,
которое будет отличатся от значения индекса последнего элемента.
'''
my_list = [56, 23, 67, 45, 67, 2, 47, 158, 31, 34, 78, 23, 78, 23, 89, 23, 36]
my_list.append('i')
a = my_list[0]
b = my_list[-2]
element_index = my_list.index(my_list[-1])
for i in range(0, element_index):
    if b < a:
        my_list.pop(element_index)
        element_index = element_index - 1
        b = my_list[element_index]
        a = my_list[0]
    if b > a:
        my_list.pop(0)
        element_index = element_index - 1
        b = my_list[element_index]
        a = my_list[0]
print('Решение №3. Значение максимального элемента списка составляет:', b)
```

Пример:

для списка ["сосиски", "горчица"] при n = 3 результат должен выглядеть так:

['сосиски$\_$1', 'горчица$\_$1', 'сосиски$\_2$', 'горчица$\_$2', 'сосиски$\_$3', 'горчица$\_$3']

```python
sample_list = ["мандаринки", "киви", "лимон"]

a1 = 'мандаринки'
a2 = 'киви'
a3 = 'лимон'
a = 0
b = 1

for i in range(3):
    a1 = 'мандаринки' + ' _ ' + str(b)
    if a1 != sample_list[0] and b == 1:
        sample_list[0] = a1
    else:
        b = int(b) + 1
        a1 = 'мандаринки' + ' _ ' + str(b)
        sample_list.append(a1)
    a2 = 'киви' + ' _ ' + str(b)
    if a2 != sample_list[1] and b == 1:
        sample_list[1] = a2
    else:
        sample_list.append(a2)
    a3 = 'лимон' + ' _ ' + str(b)
    if a3 != sample_list[2] and b == 1:
        sample_list[2] = a3
    else:
        sample_list.append(a3)
print(sample_list)
```
